```python
#Es 1 

import pandas as pd
import numpy as np
file_path = r'C:\Users\Claudia\OneDrive - Alma Mater Studiorum Università di Bologna\Documents\CLAUDIA\DataAnalyst\S6\wine.csv'
df = pd.read_csv(file_path)
print(df.head())
# Filtra i dati numerici
numeric_columns = df.select_dtypes(include=np.number)
media = numeric_columns.mean()
mediana = numeric_columns.median()
std_dev = numeric_columns.std()
moda = numeric_columns.mode()
quartili = numeric_columns.quantile([0.25, 0.5, 0.75])

print("Media:")
print(media)

print("\nMediana:")
print(mediana)

print("\nDeviazione Standard:")
print(std_dev)

print("\nModa")
print(moda)

print("\nQuartili ")
print(quartili)
```
